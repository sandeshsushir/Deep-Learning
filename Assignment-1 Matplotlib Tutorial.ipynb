```python
#create a histogram of age 
fig,ax =plt.subplots(figsize=(10,5))
df['age'].plot.hist(bins=10)
```

```python
#Create a Plot of age more than 50 and colestrol
over_50=df[df["age"]>50]
over_50
len(df),len(over_50)
```

```python
over_50
```

```python

fig,ax =plt.subplots(figsize=(10,6))
scatter= ax.scatter(x=df['age'],
                   y=df['chol'],
                   c=df['target'])

ax.set(title ="Heart disease and cholestrol level for all age",
      xlabel="age",
      ylabel="cholestrol")
ax.legend(*scatter.legend_elements(),title="target");
ax.axhline(df["chol"].mean(),
          linestyle='--')
```

```python
#crate a scatter plot with heaert disease for all age 
fig,ax =plt.subplots(figsize=(10,6))
scatter= ax.scatter(x=df['age'],
                   y=df['chol'],
                   c=df['sex'])

ax.set(title ="Heart disease and cholestrol level for all age with resp sex",
      xlabel="age",
      ylabel="cholestrol")
ax.legend(*scatter.legend_elements(),title="sex");
```

```python
#Create a scatter plot to know heart disease and cholestrol level
fig,ax =plt.subplots(figsize=(10,6))
scatter= ax.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target'])

ax.set(title ="Heart disease and cholestrol level over 50 age",
      xlabel="age",
      ylabel="cholestrol")
ax.legend(*scatter.legend_elements(),title="target");
ax.axhline(over_50["chol"].mean(),
          linestyle='--')
```

```python
#create the plot with 

fig, ax1 =plt.subplots(figsize=(10,6))
scatter =ax1.scatter(x=over_50['age'],
                   y=over_50['thalach'],
                   c=over_50['target'])
ax1.set(title="heart disease and max heart rate",
       xlabel="Age",
       ylabel="Max heart rate")
ax1.legend(*scatter.legend_elements(),title="target")
ax1.axhline(over_50['thalach'].mean(),
           linestyle="--")

```
